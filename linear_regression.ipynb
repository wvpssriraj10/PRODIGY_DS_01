```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("population_data.csv")
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
df_filtered = df[['World Development Indicators', 'Data Source']]
df_filtered = df_filtered.dropna()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df_filtered['World Development Indicators'][:10], y=df_filtered['Data Source'][:10])
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Population (2022)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Data Source'][:10], bins=20, kde=True)
plt.xticks(rotation=90)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution in Population")
plt.show()
```
